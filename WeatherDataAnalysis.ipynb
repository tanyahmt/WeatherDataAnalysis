objective: analyzing  weather of three cities of Iran

```python

import pandas as pd
import matplotlib.pyplot as plt

# creating dataframe
isfahan = pd.read_csv('/content/isfahan_weather_last_week.csv')
# print(isfahan)
# print(isfahan.columns)

qom = pd.read_csv('/content/qom_weather_last_week.csv')
# print(qom)
# print(qom.columns)

yazd = pd.read_csv('/content/yazd_weather_last_week.csv')
# print(yazd)
# print(yazd.columns)


# plotting data
plt.plot(isfahan['Date'], isfahan['Temperature (°C)'], label='isfahan')
plt.plot(qom['Date'], qom['Temperature (°C)'], label='qom')
plt.plot(yazd['Date'], yazd['Temperature (°C)'], label='yazd')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Temperature of cities')
plt.title('Temperature Trend Over 7 Days')
plt.legend()

# Show the plot
plt.show()
```

conculsion:

by looking at the plot we can clearly observe that qom and yazd had the same temperature monday and wednesday.
while yazd and isfahan had the same temperature level on sunday.

it must be mentioned that isfahan and qom had no affect on eachother.

### Qom Weather

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Open the DataFrame (assuming it's stored in a CSV file called 'isfahan_weather_last_week.csv')
df = pd.read_csv('/content/qom_weather_last_week.csv')

# Count the number of occurrences of each weather condition
weather_counts = df['Weather'].value_counts()

# Define colors for each weather condition
weather_colors = {'Sunny': 'orange', 'Rainy': 'blue', 'Partly Cloudy': 'gray', 'Cloudy': 'lightblue'}

# Create a color list based on the counts
color_list = [weather_colors[weather] for weather in weather_counts.index]

# Create a bar plot with specified colors
plt.figure(figsize=(10, 6))
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette=color_list)
plt.title('Qom Different Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Days')

# Customize legend (optional)
legend_labels = [plt.Rectangle((0,0),1,1, color=weather_colors[weather]) for weather in weather_colors]
plt.legend(legend_labels, weather_colors.keys(), loc='upper right')

plt.show()

```
